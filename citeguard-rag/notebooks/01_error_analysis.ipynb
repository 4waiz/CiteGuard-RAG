# CiteGuard-RAG: error analysis

Open this notebook **after** you have run `citeguard evaluate` so the CSV outputs exist.

```python
import json
from pathlib import Path
import pandas as pd

ROOT = Path('..').resolve()
df = pd.read_csv(ROOT / 'outputs' / 'tables' / 'claim_eval.csv')
df.head()
```

```python
df['predicted_label'].value_counts()
```

```python
has_gold = df['gold_label'].astype(str).str.len() > 0
sub = df[has_gold]
if len(sub):
    print('accuracy vs gold:', sub['is_correct'].astype(bool).mean())
    print(pd.crosstab(sub['gold_label'], sub['predicted_label']))
```

```python
errors = sub[~sub['is_correct'].astype(bool)].copy()
errors[['example_id','claim_text','gold_label','predicted_label','rationale']].head(20)
```
